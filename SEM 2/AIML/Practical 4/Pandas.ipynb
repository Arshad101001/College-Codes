## DataFrame- two dimensional, tabular format
## Method 1: dataframe from dictionary

```python
import pandas as pd
data={
    'Name':['A','B','C','D'],
    'Age':[22,22,21,23],
    'Marks':[85,92,78,95],
    'Grade':['A','A','B','A']
}
df=pd.DataFrame(data)
print(df)
type(df)
```

## Method 2: from a list of dictionary

```python
records=[
    {'Name':'Arjun','Dept':'CS','CGPA':8.5},
    {'Name':'Priya','Dept':'IT','CGPA':9.1},
    {'Name':'Rahul','Dept':'CS','CGPA':7.8}
]
df=pd.DataFrame(records)
print(df)
```

## Method 3: from numpy array

```python
import numpy as np
arr=np.array([[1,85,22],
            [2,92,25],
            [3,78,21]])
df=pd.DataFrame(arr,columns=['Rollno','Marks','Age'],index=["Row1","Row2","Row3"])
print(df)
```

```python
df.to_csv('output.csv',index=False)
```
