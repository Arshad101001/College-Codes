## 1. Creating arrays from existing data - np.array()

```python
import numpy as np
```

```python
# 1D array from list
arr1=np.array([10,20,30,40])
print(arr1)
```

```python
#print data type of array
print(arr1.dtype)
```

```python
#2d array(matrix)
arr2=np.array([[1,2,3],[4,5,6]])
print(arr2)
print(arr2.shape)
```

## 2.Constant field array - np.zeros()-array of all zeros

```python
#np.zeros(rowSize,colSize) #1d zeros
z1=np.zeros(5)
print(z1)
#2d zeros
newArray=np.zeros((2,3))
print(newArray)
#integer zeros
z2=np.zeros((2,3),dtype=int)
print(z2)
```

## np.ones()-array of all ones

```python
#np.ones(rowSize,colSize) #1d zeros
z3=np.ones((2,3))
print(z3)
```

## np.full() - custom fill value - fills an array of any shape

```python
#np.full(size,constValue)
f1=np.full(5,7)
print(f1)
f2=np.full((3,3),(3.14))
print(f2)
```

## Range - based arrays
## np.arrange() - works line range()

```python
#arange(stopValue)
ArangeArray=np.arange(6)# if we are only value it will consider stop value
print(ArangeArray)
```

```python
#arange(start,stop)
print(np.arange(1,6))
#arange(start,stop,step)-we can use float also
print(np.arange(0,10,2))
```

## np.linspace()- it equally spaced and create in it

```python
# 5 values from 0 to 1
print(np.linspace(0,1,5))

# 4 values from 0 to 10
print(np.linspace(0,10,4))

# exclude endpoint
print(np.linspace(0,1,5,endpoint=False))
```

## 5. Random arrays- numpy.random
## numpy's random module provides function to create arrays
## np.random.rand()- Uniform[0, 1]

```python
#1d.5 val;ues uniform between 0 and 1
print(np.random.rand(5))
```

```python
#2d: 2x3 matrix
print(np.random.rand(2,3))
```

```python
#np.random.randint() - random integers
# 0 to 9
print(np.random.randint(0,10,size=5))
```

```python
#2d matrix
print(np.random.randint(1,100,size=(3,3)))
```

```python
#uniform b/w 2 and 8
np.random.uniform(2,8,size=5)
```

```python
a=np.array([[3,6],[9,12]])
# zeros_like- same shape, all zeros
print(np.zeros_like(a))
```

```python
print(np.ones_like(a))

print(np.full_like(a,fill_value=89))
```

```python
# create one 2d array of(2,3) size
a2=np.array([[1,2,3],[4,5,6]])
print(a2)
#find size, datatype,dimension,shape of the created array
print(a2.size)
print(a2.dtype)
print(a2.shape)
print(a2.ndim)
```
