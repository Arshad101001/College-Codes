<h3>Implementation of perceptron algo for OR function. </h3>

```python
import numpy as np

class per:
    def __init__(self, alpha=0.1, n_itr=20):
        self.alpha = alpha
        self.n_itr = n_itr
        self.w = None
        self.b = None

    def act_fun(self, x):
        return np.where(x >= 0, 1, 0)

    def fit(self, X, y):
        n_sam, n_feat = X.shape
        self.w = np.zeros(n_feat)
        self.b = 0

        y_ = np.array([1 if i > 0 else 0 for i in y])

        for epoch in range(self.n_itr):
            epoch_err = 0
            for idx, x_i in enumerate(X):
                net_in = np.dot(x_i, self.w) + self.b
                y_pred = self.act_fun(net_in)

                err = y_[idx] - y_pred
                up = self.alpha * err

                self.w += up * x_i
                self.b += up

                epoch_err += int(err != 0)

            if epoch_err == 0:
                print(f"Converged at epoch {epoch + 1}")
                break

    def predict(self, X):
        net_in = np.dot(X, self.w) + self.b
        return self.act_fun(net_in)

    def res(self, X, y):
        pred = self.predict(X)
        print("Actual:     ", y)
        print("Predicted:  ", pred)

        
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,1])

p = per(alpha=0.1, n_itr=20)
p.fit(X, y)
p.res(X, y)
```
