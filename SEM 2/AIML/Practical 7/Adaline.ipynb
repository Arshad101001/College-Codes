<h3>Implementation of adaline algo for and function. </h3>

```python
import numpy as np

class adaline:
    def __init__(self,in_size, alpha=0.1, ep=100):
        self.alpha = alpha
        self.ep = ep
        self.w = np.zeros(in_size)
        self.b = 0
        
    def act_fun(self, X):
        return X

    def pred(self,X):
        return self.act_fun(np.dot(X,self.w)+self.b)

    def train(self,X,y):
        for e in range(self.ep):
            for i in range(len(X)):
                predi=self.pred(X[i])
                err=y[i]-predi
                #update weights and bias
                self.w+=self.alpha*err*X[i]
                self.b+=self.alpha*err

    def eval(self,X):
        return np.where(self.pred(X)>=0.5,1,0)

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

a = adaline(in_size=2,alpha=0.1, ep=100)
a.train(X,y)
predictions=a.eval(X)
print(predictions)
```

```python
for i, p in enumerate(predictions):
    print(f"Input:{X[i]} => Predicted:{p}=>Actual:{y[i]}")
```
